```python
class Reverse:
    def reverse_number(self, num):
        rev = 0
        while num > 0:
            rem = num % 10
            rev = rev * 10 + rem
            num //= 10
        return rev

num = int(input("Enter the number: "))
rev_obj = Reverse()
result = rev_obj.reverse_number(num)
print(f"The reverse of {num} is {result}")
```
