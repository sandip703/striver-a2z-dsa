# Basic Maths - Topic 1: Digit Concept & Extraction of Digits

## 📚 THEORY & CONCEPT

### What is Digit Extraction?
Digit extraction means getting each individual digit from a number separately.

### The Magic Formula
- **To get last digit**: `number % 10` (modulo operator gives remainder)
- **To remove last digit**: `number / 10` (integer division)

### Why does modulo 10 give last digit?
Numbers divisible by 10 always end with 0. When you divide any number by 10, the remainder is always the last digit!

**Example**: 
```
7789 ÷ 10 = 778 (quotient) with remainder 9
So: 7789 % 10 = 9 ✓
```

### Visual Process - Extracting digits from 7789:

```
Step 1: 7789 % 10 = 9  (get last digit)
        Then: 7789 / 10 = 778

Step 2: 778 % 10 = 8   (get next digit)
        Then: 778 / 10 = 77

Step 3: 77 % 10 = 7
        Then: 77 / 10 = 7

Step 4: 7 % 10 = 7
        Then: 7 / 10 = 0 (STOP - number becomes 0)
```

### Flow Diagram:
```
    START
      │
```

**Digits extracted in order: 9, 8, 7, 7 (reverse order!)**

## 📝 PSEUDOCODE

```
PROCEDURE extract_digits(N):
    WHILE N > 0 DO
        last_digit = N % 10
        PRINT last_digit
        N = N / 10
    END WHILE
END PROCEDURE
```

### Step-by-step Explanation:
| Step | What happens | Why |
|------|-------------|-----|
| 1 | Check if N > 0 | Continue until no digits left |
| 2 | Get N % 10 | Extract the last digit |
| 3 | Print/Store digit | Use the digit as needed |
| 4 | Set N = N / 10 | Remove the last digit |
| 5 | Repeat from step 1 | Process remaining digits |

## 💻 PYTHON CODE

```python
def extract_digits(n):
    """Extract and print all digits of a number"""
    print(f"Extracting digits from {n}:")
    
    while n > 0:
        last_digit = n % 10
        print(f"Digit extracted: {last_digit}")
        n = n // 10  # Integer division to remove last digit
    
    print("All digits extracted!")

def extract_and_store(n):
    """Extract digits and store in a list"""
    digits = []
    while n > 0:
        digits.append(n % 10)
        n = n // 10
    return digits  # Returns digits in reverse order

# Example usage
number = 7789
extract_digits(number)

# Store for later use
digits_list = extract_and_store(7789)
print(f"\nStored digits (reverse order): {digits_list}")
```

## ⭐ Key Takeaways

| Concept | Formula | Time Complexity |
|---------|---------|-----------------|
| Get last digit | `N % 10` | O(1) |
| Remove last digit | `N // 10` | O(1) |
| Extract all digits | Loop until N = 0 | O(log₁₀ N) |

> **Why O(log₁₀ N)?** Because we divide by 10 each time. Number of iterations = number of digits = floor(log₁₀ N) + 1

Count digits in a number


50

Problem Statement: Given an integer N, return the number of digits in N.

Examples
Example 1:
Input:N = 12345
Output:5
Explanation:  The number 12345 has 5 digits.
                        
Example 2:
Input:N = 7789              
Output: 4
Explanation: The number 7789 has 4 digits. 

### BruteForceApproach

````python
class Solution:
    def count(self,num):
        count=0
        while num > 0:
            count+=1
            num=num//10  
        
        return count

obj=Solution()
print(obj.count(123456))
````

### Optimized Approach

````python
import math
class Solution:
    def count(self,num):
        return int(math.log10(num) + 1)

obj=Solution()
print(obj.count(123456))
````
