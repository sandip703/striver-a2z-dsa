# Basic Maths - Topic 2: Count Digits

## 📚 THEORY & CONCEPT

### Problem Statement
Given a number **N**, count how many digits are present in that number.

**Examples:**
| Input | Output | Why? |
|-------|--------|------|
| 156 | 3 | Digits: 1, 5, 6 |
| 7789 | 4 | Digits: 7, 7, 8, 9 |
| 7 | 1 | Only one digit |

### The Intuition
Using digit extraction concept, each time we remove the last digit, we count one digit. When the number becomes 0, the count equals total digits!

### Visual Representation:

```
Number: 7789
┌─────┬──────┬──────────┬─────────┬─────────┐
│Step │ N    │ N % 10   │ N = N/10│ Count   │
├─────┼──────┼──────────┼─────────┼─────────┤
│ 1   │ 7789 │ (extract)│ 778     │ 1       │
│ 2   │ 778  │ (extract)│ 77      │ 2       │
│ 3   │ 77   │ (extract)│ 7       │ 3       │
│ 4   │ 7    │ (extract)│ 0       │ 4       │
│ 5   │ 0    │ STOP     │ -       │ -       │
└─────┴──────┴──────────┴─────────┴─────────┘

Final Count = 4 ✓
```

### Alternative Approach: Mathematical Formula
Using logarithm base 10:
```
Number of digits = ⌊log₁₀(N)⌋ + 1

Example: 7789
log₁₀(7789) ≈ 3.89
floor(3.89) = 3
3 + 1 = 4 digits ✓
```

## 📝 PSEUDOCODE

### Method 1: Iterative (Extraction Method)
```
PROCEDURE count_digits_extraction(N):
    count = 0
    WHILE N > 0 DO
        count = count + 1
        N = N / 10
    END WHILE
    RETURN count
END PROCEDURE
```

### Method 2: Mathematical (Logarithm Method)
```
PROCEDURE count_digits_log(N):
    IF N == 0 THEN
        RETURN 1
    END IF
    RETURN floor(log10(N)) + 1
END PROCEDURE
```

### Step-by-step Explanation (Extraction Method):

| Step | Operation | Explanation |
|------|-----------|-------------|
| 1 | Initialize `count = 0` | Start with zero digits counted |
| 2 | Check `N > 0` | Continue if digits remain |
| 3 | `count = count + 1` | One more digit found |
| 4 | `N = N // 10` | Remove the last digit |
| 5 | Repeat from step 2 | Process remaining digits |
| 6 | Return `count` | Final digit count |

## 💻 PYTHON CODE

```python
import math

def count_digits_iterative(n):
    """
    Method 1: Count digits by repeatedly dividing by 10
    Time Complexity: O(log₁₀ n)
    Space Complexity: O(1)
    """
    # Handle edge case: N = 0
    if n == 0:
        return 1
    
    count = 0
    original = n  # Store for display
    
    while n > 0:
        count += 1
        print(f"Step {count}: {n} → remove last digit → {n // 10}")
        n = n // 10
    
    print(f"\nNumber {original} has {count} digits")
    return count


def count_digits_logarithmic(n):
    """
    Method 2: Count digits using logarithm formula
    Formula: floor(log₁₀(N)) + 1
    Time Complexity: O(1)
    Space Complexity: O(1)
    """
    if n == 0:
        return 1
    
    digits = math.floor(math.log10(n)) + 1
    return digits


def count_digits_string(n):
    """
    Method 3: Using string conversion (for understanding only)
    Not recommended for large numbers in interviews
    """
    return len(str(abs(n)))


# Example usage with multiple test cases
test_numbers = [7789, 156, 7, 0, 12345, 1000]

print("=" * 50)
print("COUNTING DIGITS - MULTIPLE METHODS")
print("=" * 50)

for num in test_numbers:
    print(f"\n{'─' * 40}")
    print(f"Number: {num}")
    print(f"Method 1 (Iterative): {count_digits_iterative(num)} digits")
    print(f"Method 2 (Logarithm):  {count_digits_logarithmic(num)} digits")
    print(f"Method 3 (String):     {count_digits_string(num)} digits")
```

## ⏱️ Complexity Analysis

| Method | Time Complexity | Space Complexity | Best For |
|--------|----------------|------------------|----------|
| Iterative (Division) | O(log₁₀ n) | O(1) | Understanding concept |
| Logarithmic | O(1) | O(1) | Quick calculations |
| String conversion | O(log₁₀ n) | O(log₁₀ n) | Simple code (not recommended for interviews) |

### Why O(log₁₀ n)?
- Each division by 10 reduces the number
- Number of iterations = number of digits
- For number N with d digits: 10^(d-1) ≤ N < 10^d
- Therefore d ≈ log₁₀ N

---

## ⚠️ Important Notes

1. **Handle N = 0 separately** - The while loop condition (N > 0) won't execute, so 0 would return 0 digits if not handled
2. **Negative numbers** - Take absolute value first: `abs(N)`
3. **Very large numbers** - Logarithmic method might have precision issues; iterative method is safer

Problem Statement: Given an integer N, return the number of digits in N.

Examples
Example 1:
Input:N = 12345
Output:5
Explanation:  The number 12345 has 5 digits.
                        
Example 2:
Input:N = 7789              
Output: 4
Explanation: The number 7789 has 4 digits. 

### BruteForceApproach

````python
class Solution:
    def count(self,num):
        count=0
        while num > 0:
            count+=1
            num=num//10  
        
        return count

obj=Solution()
print(obj.count(123456))
````

### Optimized Approach

````python
import math
class Solution:
    def count(self,num):
        return int(math.log10(num) + 1)

obj=Solution()
print(obj.count(123456))
````
