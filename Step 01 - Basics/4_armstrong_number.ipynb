```python
class Armstrong:
    def check_armstrong(self, num):
        duplicate = num
        sum_of_cubes = 0
        while num > 0:
            rem = num % 10
            sum_of_cubes += rem ** 3
            num = num // 10
        if sum_of_cubes == duplicate:
            print(f"The number {duplicate} is an Armstrong number")
        else:
            print(f"The number {duplicate} is not an Armstrong number")

obj = Armstrong()
num = int(input("Enter the number: "))
obj.check_armstrong(num)
```
