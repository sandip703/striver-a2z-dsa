### if the number of the test is same from the forward and backward direction then the number  is the plaindrome

```python



class plaindrome:
 def checkPlaindrome(self,num):
    duplicate=num
    rev=0
    while num>0:
        rem=num%10
        rev=rev*10+rem
        num=num//10

    if duplicate==rev:
        print(f"the number {duplicate} is plaindrome")
    else:
        print(f"the number {duplicate} is not plaindrome")    
obj=plaindrome()

num=int(input("enter the number to be checked for the plaindrome :"))
obj.checkPlaindrome(num)


```
