```python
class Palindrome:
    def check_palindrome(self, num):
        duplicate = num
        rev = 0
        while num > 0:
            rem = num % 10
            rev = rev * 10 + rem
            num = num // 10

        if duplicate == rev:
            print(f"The number {duplicate} is palindrome")
        else:
            print(f"The number {duplicate} is not palindrome")

obj = Palindrome()
num = int(input("Enter the number to be checked for palindrome: "))
obj.check_palindrome(num)
```
